```python
import numpy as np
import scipy as sp
import scipy.signal as sig
import matplotlib.pyplot as plt
from astropy.io import fits
import astropy.table as table
import matplotlib as mpl
import matplotlib.image as mpimg
from matplotlib.colors import LogNorm
from sklearn.mixture import GMM

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams['figure.figsize'] = 8, 6
```

```python
im = fits.open('../NGC2558/NGC2558.disp_cor.fits')[0].data

plt.figure(figsize = (12, 12), dpi = 250)
allfibers = plt.imshow((im.T/np.average(im, axis = 1).T).T, origin = 'lower', aspect = 1700/72, cmap = 'gnuplot2', norm = LogNorm(), interpolation = 'nearest')
plt.colorbar(allfibers, shrink = 0.8)
skyfibers = [80, 2, 22, 70, 54, 37]
for fiber in skyfibers:
    plt.axhline(float(fiber), linestyle = '--', c = 'b')
plt.show()

plt.figure(figsize = (12, 12), dpi = 250)
allfibers = plt.imshow((im[:,1100:1300].T/np.average(im[:,1100:1300], axis = 1).T).T, origin = 'lower', extent = [1100, 1300, 0, 81], aspect = 200/72, cmap = 'gnuplot2', norm = LogNorm(), interpolation = 'nearest')
plt.colorbar(allfibers, shrink = 0.8)
skyfibers = [80, 2, 22, 70, 54, 37]
for fiber in skyfibers:
    plt.axhline(float(fiber), linestyle = '--', c = 'g')
plt.show()
```
